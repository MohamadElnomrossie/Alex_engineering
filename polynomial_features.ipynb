```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.pipeline import Pipeline
```

```python
X,y=load_breast_cancer(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
transformer=PolynomialFeatures()
```

```python
transformer.fit(X_train)
X_train=transformer.transform(X_train)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
transformer.get_feature_names_out()
```

```python
logisitc=LogisticRegression(max_iter=4000)
```

```python
logisitc.fit(X_train,y_train)
```

```python
logisitc.predict(X_test)
```

```python
transfomred_X_test=transformer.transform(X_test)
```

```python
transfomred_X_test.shape
```

```python
predicted=logisitc.predict(transfomred_X_test)
```

```python
print(classification_report(y_test,predicted))
```

```python
pipe=Pipeline([('transformer',PolynomialFeatures()),
              ('logistic',LogisticRegression(max_iter=5000))])
```

```python
X,y=load_breast_cancer(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
pipe.fit(X_train,y_train)
```

```python
predicted=pipe.predict(X_test)
```

```python
print(classification_report(y_test,predicted))
```
